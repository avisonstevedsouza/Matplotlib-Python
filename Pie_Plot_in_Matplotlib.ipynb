```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
classes=['Physics','Chemistry','Maths','Science','SST']
marks=[89,45,78,23,90]
```

```python
plt.pie(marks,labels=classes)
plt.show()
```

```python
colors=['red','green','blue','orange','pink']
plt.pie(marks,labels=classes,colors=colors)
plt.show()
```

```python
colors=['red','green','blue','orange','pink']
plt.pie(marks,labels=classes,
        colors=colors,
        autopct='%0.2f%%')
plt.show()
```

```python
colors=['red','green','blue','orange','pink']
plt.pie(marks,labels=classes,
        colors=colors,
        autopct='%0.2f%%')
plt.show()
```

```python
explode_values=[0.1,0,0,0,0]
colors=['red','green','blue','orange','pink']
plt.pie(marks,labels=classes,
        colors=colors,
        autopct='%0.2f%%',
        explode=explode_values)
plt.show()
```

```python
explode_values=[1,1,1,1,1]
colors=['red','green','blue','orange','pink']
plt.pie(marks,labels=classes,
        colors=colors,
        autopct='%0.2f%%',
        explode=explode_values)
plt.show()
```

```python
explode_values=[0.1,0,0,0,0]
colors=['red','green','blue','orange','pink']
plt.pie(marks,labels=classes,
        colors=colors,
        autopct='%0.2f%%',
        explode=explode_values,
        shadow=True)
plt.show()
```

```python
explode_values=[0.1,0,0,0,0]
colors=['red','green','blue','orange','pink']
plt.pie(marks,
        labels=classes,
        colors=colors,
        autopct='%0.2f%%',
        explode=explode_values,
        shadow=True,
        radius=1.6)
plt.show()
```

```python
textprops={'fontsize':14,
           'color':'white'}
wedgeprops={'linewidth':3,
            'linestyle':'--',
            'edgecolor':'White'}
explode_values=[0.1,0,0,0,0]
plt.pie(marks,
        labels=classes,
        colors=colors,
        autopct='%0.2f%%',
        explode=explode_values,
        shadow=True,
        radius=1.6,
        textprops=textprops,
        wedgeprops=wedgeprops)
plt.legend()
plt.show()
```

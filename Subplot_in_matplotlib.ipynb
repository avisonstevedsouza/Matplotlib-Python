```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
plt.plot(study_hours,marks,'r--')
plt.show()
```

```python
plt.hist(marks)
plt.show()
```

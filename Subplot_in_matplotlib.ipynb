```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
study_hours=[2,3,4,4,5,6,7,7,8,9,9,10,11,11,12]
marks=[6,10,15,20,34,44,55,60,55,67,70,80,90,99,100]
```

```python
plt.scatter(study_hours,marks)
plt.show()
```

```python
plt.plot(study_hours,marks,'r--')
plt.show()
```

```python
plt.hist(marks)
plt.show()
```

```python
plt.plot(study_hours,marks,'r--')
plt.plot(study_hours,marks,'bo')
plt.show()
```

# SubPlots

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.scatter(study_hours,marks)

plt.subplot(2,2,2)
plt.plot(study_hours,marks,'r--')

plt.subplot(2,2,3)
plt.hist(marks)

plt.subplot(2,2,4)
plt.plot(study_hours,marks,'r--')
plt.plot(study_hours,marks,'bo')

plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.scatter(study_hours,marks,color='red',alpha=0.5)

plt.subplot(2,2,2)
plt.plot(study_hours,marks,'r--')

plt.subplot(2,2,3)
plt.hist(marks)

plt.subplot(2,2,4)
plt.plot(study_hours,marks,'r--')
plt.plot(study_hours,marks,'bo')

plt.show()
```

# Saving To Local Machine

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.scatter(study_hours,marks,color='red',alpha=0.5)


plt.subplot(2,2,2)
plt.plot(study_hours,marks,'r-.')


plt.subplot(2,2,3)
plt.hist(marks,color='red',histtype='step')


plt.subplot(2,2,4)
plt.plot(study_hours,marks,'r--')
plt.plot(study_hours,marks,'ro')

plt.savefig('/content/sample_data/data.png',facecolor='k')

plt.show()
```

```python
plt.figure(figsize=(12,12))

plt.subplot(2,2,1)
plt.scatter(study_hours,marks,color='red',alpha=0.5)
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title("Study Hours vs Marks")

plt.subplot(2,2,2)
plt.plot(study_hours,marks,'r-.')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title("Study Hours vs Marks")

plt.subplot(2,2,3)
plt.hist(marks,color='red',histtype='step')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title("Study Hours vs Marks")

plt.subplot(2,2,4)
plt.plot(study_hours,marks,'r--')
plt.plot(study_hours,marks,'ro')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title("Study Hours vs Marks")

plt.savefig('/content/sample_data/full_data.png',facecolor='k')

plt.show()
```
